```python
import pandas as pd

# Load the training set
training_set = pd.read_csv('artifacts/training_set.csv')

# Display basic information about the dataset
print(f"Dataset shape: {training_set.shape}")
print(f"\nColumn names: {list(training_set.columns)}")
print(f"\nFirst few rows:")
training_set.head()
```

# 1. Target Distribution


```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Target Distribution
sns.countplot(x='defaulted_within_90d', data=training_set)
plt.title('Distribution of Defaults')

```

# 2. Income vs Default (using engineered total_credit)


```python
# 2. Income vs Default (using engineered total_credit)
sns.boxplot(x='defaulted_within_90d', y='total_credit', data=training_set)
plt.title('Income (Total Credit) by Default Status')

```

# 3. Categorical Impact (Proportion of those with 'rent' tag)


```python
# 3. Categorical Impact (Proportion of those with 'rent' tag)
sns.barplot(x='defaulted_within_90d', y='flag_rent_mortgage', data=training_set)
plt.title('Proportion of Customers with Rent Payments by Default Status')
```
